## Agregação

````python
#group_city = quatromil.groupby(["city"])["total"].agg("mean")
group_city = quatromil.groupby(["city"])["total"].agg("mean").reset_index()
group_city
````

````python
group_bath_room_price = quatromil.groupby(["bathroom","rooms"])["total"].agg("mean").reset_index()
group_bath_room_price
````

````python
sns.barplot(data=group_bath_room_price, x="rooms", y="total", color="blue");
````

````python
ax = sns.barplot(data=group_bath_room_price, x="rooms", y="total", color="blue")
ax.set(xlabel="qtd quartos", ylabel = "Média Preço" );
````

````python
ax = sns.barplot(data=group_bath_room_price, x="bathroom", y="total", color="green")
ax.set(xlabel="qtd banheiros", ylabel = "Média Preço" );
````
