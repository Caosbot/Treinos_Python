## Acertos

### **Colocar o mesmo nome, trocando APP por Aplicativo**

```python
#Colocar o mesmo nome, trocando APP por Aplicativo
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

### Verificao

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

### Replace Nome_Departamento

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(" ", "_")
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

## Tratando Nulos

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_dropNull = vendas.dropna()
vendas_dropNull.info()
```

```python
#vendas['estado'] = vendas['estado'].fillna("MS")
vendas["estado"].fillna("MS",inplace=True )
vendas.info()
```

```python
vendas[vendas['Preo'].isnull()]
```

```python
print(vendas['Preo_com_frete'].mean())
```

```python
meanpreco = vendas['Preo'].mean()
vendas['Preo'].fillna(meanpreco, inplace=True)
#vendas['Preo'].fillna(vendas['Preo_com_frete'].mean(), inplace=True)
```

```python
print(meanpreco)
```

```python
vendas.info()
```

## Ajustando dados

```python
vendas[["Preo", "Preo_com_frete"]].describe()
```

```python
frete_correto = vendas.query("Preo < Preo_com_frete")
frete_correto.describe()
```

```python
frete_correto = frete_correto.reset_index()
frete_correto
```

```python
print(f"vendas = {vendas['idcompra'].count()}\nfrete_correto =  {frete_correto['idcompra'].count()}\nDiferena = {vendas['idcompra'].count()-frete_correto['idcompra'].count()}")
```

```python
frete_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

## Cumprindo Mtricas

### Departamento com mais vendas

```python
frete_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False)
```

### Mdia de Preo com frete por departamento

```python
round(frete_correto.groupby("Nome_Departamento")["Preo_com_frete"].agg("mean").sort_values(ascending=False),2)
```

```python
frete_correto.groupby("Nome_Departamento")["Preo_com_frete"].agg("mean").sort_values(ascending=False).round(2).apply(lambda x: f'R$ {x:<15,.2f}')
#resultado = resultado.round(2).apply(lambda x: f'R$ {x:<15,.2f}')

#print(resultado)
```

### Quantidade de vendas por Ms

```python
frete_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False)
```

```python
frete_correto["mes"] = frete_correto["Data"].dt.month_name()#dt.month
frete_correto
```

```python
frete_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False)
```

## Novos dados Cliente
Merge com varejo

```python
cliente = pd.read_excel("/content/cliente_varejo.xlsx")
```

```python
cliente
```

```python
cliente[["idade","renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
vendas_cliente = frete_correto.merge(cliente, how="left", on= "cliente_Log")
vendas_cliente
```

```python
idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
idcanal_renda
```

```python
type(idcanal_renda)
```

```python
idade_bandeira = round(vendas_cliente.groupby('bandeira')["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
idade_bandeira
```
